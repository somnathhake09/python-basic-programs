```python
#set A
#symmetrical & palindrome
s=input("enter a string:")

if s == s[::-1]:
    print("palindrome")
else:
    print("not palindrome")

if s[:len(s)//2]==s[len(s)//2:]:
    print("symmetrical")
else:
    print("not symmetrical")
```

```python
#prime or not
def is_prime(n):
    if n<2:
        return False
    for i in range(2,n):
        if n%i==0:
         return False
    return True
num = int(input("enter a number:"))
if is_prime(num):
    print("prime")
else:
    print("not prime")
```

```python
#perfect number
def is_perfect(n):
    s=0
    for i in range(1,n):
        if n % i == 0:
            s+=1
            return s == n
num=int(input("enter a number:"))
if is_perfect(num):
    print("perfect number")
else:
    print("not perfect number")
    
```
